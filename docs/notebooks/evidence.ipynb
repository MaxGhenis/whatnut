# Evidence Review

This appendix summarizes the primary sources underlying our QALY estimates. All sources include DOIs for verification.

```python
from whatnut.evidence import SOURCES
import pandas as pd
```

## Primary Sources

Our analysis draws on meta-analyses, cohort studies, and randomized controlled trials.

```python
source_data = []
for s in SOURCES:
    source_data.append({
        'Study': s.id,
        'Type': s.study_type,
        'Sample Size': f"{s.sample_size:,}" if s.sample_size else "â€”",
        'Key Finding': s.key_finding,
    })

pd.DataFrame(source_data).style.hide(axis='index')
```

## Effect Sizes

The following table shows mortality effect sizes from the meta-analyses.

```python
effect_data = []
for s in SOURCES:
    if s.effect_size is not None:
        effect_data.append({
            'Source': s.id,
            'Metric': s.effect_size.metric,
            'Estimate': f"{s.effect_size.point_estimate:.2f}",
            '95% CI': f"[{s.effect_size.ci_lower:.2f}, {s.effect_size.ci_upper:.2f}]",
        })

pd.DataFrame(effect_data).style.hide(axis='index')
```
